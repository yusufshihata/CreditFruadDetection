```python
# We won't need the Unnamed, ID columns
df.drop(['Unnamed: 0', 'ID'], axis=1, inplace=True)
```

```python
# We need to get outliers from the data

numerical_fields = ['AGE', 'YEARS_EMPLOYED', 'INCOME', 'NO_OF_CHILD']

for field in numerical_fields:
    plt.figure(figsize=(10, 10))
    sns.boxplot(df, y=field)
    plt.show()
```

```python
# We need to do encoding on categorical data
from sklearn.preprocessing import OrdinalEncoder

one_hot_cols = ['INCOME_TYPE', 'FAMILY_TYPE', 'HOUSE_TYPE', 'GENDER', 'CAR', 'REALITY']

df = pd.get_dummies(df, columns=one_hot_cols).replace({True: 1, False: 0})

ordinalEn = OrdinalEncoder()

df['EDUCATION_TYPE'] = ordinalEn.fit_transform(df[['EDUCATION_TYPE']])
```

```python
df.info()
```

```python
# Log Transformation for the income
df['INCOME'] = np.log1p(df['INCOME'])
```

```python
df.head(5)
```

```python
def cut_outliers(*features):
    for feature in features:
        Q1 = df[feature].quantile(0.25)
        Q3 = df[feature].quantile(0.75)
        IQR = Q3 - Q1

        lower = Q1 - IQR * 1.5
        upper = Q3 + IQR * 1.5

        df[feature] = df[feature].clip(lower=lower, upper=upper)
```

```python
cut_outliers('INCOME', 'AGE', 'NO_OF_CHILD', 'YEARS_EMPLOYED', 'FAMILY SIZE')
```

```python
numerical_fields = ['AGE', 'YEARS_EMPLOYED', 'INCOME', 'NO_OF_CHILD']

for field in numerical_fields:
    plt.figure(figsize=(10, 10))
    sns.boxplot(df, y=field)
    plt.show()
```

```python
df.head(5)
```

```python
# We need to scale some features (INCOME, AGE, BEGIN_MONTH, NO_OF_CHILD)
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

features = ['INCOME', 'AGE', 'BEGIN_MONTH', 'NO_OF_CHILD', 'FAMILY SIZE']

for feature in features:
    df[feature] = scaler.fit_transform(df[[feature]])
```

```python
df.head()
```

```python
df.to_csv("credit_dataset_preprocessed.csv")
```

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
import numpy as np

# Assuming df is already defined
X = df.drop("TARGET", axis=1)
y = df["TARGET"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Convert to DMatrix format
D_train = xgb.DMatrix(X_train, label=y_train)
D_test = xgb.DMatrix(X_test, label=y_test)
```

```python
from imblearn.under_sampling import RandomUnderSampler

undersample = RandomUnderSampler(sampling_strategy='auto', random_state=42)
X_resampled, y_resampled = undersample.fit_resample(X_train, y_train)

from collections import Counter

# Calculate the class imbalance ratio
counter = Counter(y_train)
scale_pos_weight = counter[0] / counter[1] 
```

```python
# Train the model
model = xgb.XGBClassifier(scale_pos_weight=scale_pos_weight)

model.fit(X_resampled, y_resampled)

train_pred = model.predict(X_resampled)

# Convert probabilities to class labels
train_pred_labels = np.argmax(train_pred)

# Calculate training accuracy
train_accuracy = accuracy_score(y_resampled, train_pred)
print(f"Training Accuracy: {train_accuracy}")

# Predict probabilities
pred = model.predict(X_test)

# Convert probabilities to class labels (index of the maximum probability)
pred_labels = np.argmax(pred)
```

```python
# We need a high recall of 1 which means that it identify most of fraud, if we achieve this we can just stop as this is the most dengrous thing 
print(classification_report(y_test, pred))
```
